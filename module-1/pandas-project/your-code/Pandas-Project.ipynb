```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 1000)
#importing the libraries we'll most likely need
```

```python
sharks = pd.read_csv('../GSAF5.csv')
#importing the file we need
sharks.head()
```

```python
print(len(sharks),'\n')
#calculated the length to see how many values we have
null_cols = sharks.isnull().sum()
print(null_cols[null_cols > 0])
#checking which columns contain too many null values compared to the total amount of values
```

```python
drop_cols = list(null_cols[null_cols > 3000].index)
print(drop_cols)
sharks = sharks.drop(drop_cols, axis=1)
sharks = sharks.drop(columns = ['original order'])
sharks
#Dropped unnecessary columns such as the the two unnamed ones and the time since more than 
#half the dataset didn't have a value for it
```

```python
sharks = sharks.fillna('N/A')
sharks
```

```python
print(sharks.dtypes)
#Types are clean for the most part
sharks.head(100)
```

```python
sharks['Date'] = sharks['Date'].str.replace('Reported','')
sharks[sharks['Date'].str.contains('Reported')]
```

```python
sharks['Name'] = sharks['Name'].drop_duplicates(keep=False)
sharks['Name'] = sharks['Name'].fillna('N/A')
sharks['Name'].head()
```

```python
sharks.columns = (sharks.columns.str.replace(' ','')
                     .str.replace('\W',''))
sharks.columns
```

```python
# sharks['Species'][0:74]
#Checking values
```

```python
drop_rows = list(sharks[sharks['Species'].str.contains('involvement')==True].index)
print(drop_rows)
sharks = sharks.drop(sharks.index[drop_rows])
sharks[sharks['Species'].str.contains('involvement')]
```

```python
print('Amount of Cases:',len(sharks))
print('Duplicate CN2:',len(sharks[sharks['CaseNumber1']==sharks['CaseNumber2']]))
sharks = sharks.drop('CaseNumber2', axis=1)
sharks
```

```python
sharks['FatalYN'].unique()
x = list(sharks['FatalYN'])
y = []
for i in range(len(x)):
    if x[i] == 'N'or'n'or'N 'or' N':
        y.append(False)
    elif x[i] == 'Y':
        y.append(True)
    else:
        y.append('Unsure')
sharks['FatalYN'] = y
sharks.to_csv('../your-code/GSAF5.csv', index=False)
```
