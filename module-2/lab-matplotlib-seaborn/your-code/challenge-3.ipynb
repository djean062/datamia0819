#### Use the appropriate plot to visualize the column `Gender`. There are 2 ways of doing it. Do it both ways.

*Hint: Use matplotlib and seaborn*

```python
# your code here- 1st way
# titanic.Gender.plot(kind ='count',linewidth=2)
labels = ['male', 'female']
explode = (0, 0.05)
sizes = [titanic[titanic['Gender'] == labels[0]].shape[0] / titanic.shape[0] * 100, titanic[titanic['Gender'] == labels[1]].shape[0] / titanic.shape[0] * 100]
plt.pie(sizes, labels = labels, autopct = '%1.1f%%', explode = explode, startangle = 90)
```

```python
# your code here- 2nd way
sns.swarmplot(x = 'Gender', kind = 'count', data = titanic)
```

#### Use the appropriate plot to visualize the column `Pclass` 


```python
# your code here
Pclass_count  = titanic['Pclass'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(Pclass_count.index, Pclass_count.values,palette='husl')
plt.title('Pclass')
plt.ylabel('Passengers', fontsize=12)
plt.xlabel('Classes', fontsize=12)
plt.show()
```
